<p>Attention Mechanisms and Positional Encoding</p>
<p>- They resolved the limitations of RNNs, which processed sequences like "how was your day?" which processed tokens (1 per word) but failed to capture long-range relationships</p>
<p>- Attention mechanisms require positional information for each token in a sequence.</p>
<p>- Given a token 'e', a vector is created o describe the token's position in a sequence using cosine and sine (contains position information needed by attention mechanisms)</p>
<p>Self Attention - Used by transformers which weighs the attention of all tokens in the sequence simultaneously, capturing long range dependencies.</p>
<p>Positional encoder: includes sequence order information into the models</p>
<p>&nbsp;</p>
<p>Positional encoder class:</p>
<p>to create this, we subclass PyTorch Module class and edit the init method, setting in the max sequence length and the embedding dimensional for the transformer model that it will later belong to.</p>
<p>Also create a matrix, pe, for positional encoding of sequences up to our specified maximum length.</p>
<p>Initially, it contains values representing the position indices of tokens in the sequence.</p>
<p>Position indices are scaled using a factor stored in the div_term varaible and combined with sine and consine functions.</p>

```python
# Transformer Architecture

# Subclass an appropriate PyTorch class 
class PositionalEncoder(nn.Module):
    def __init__(self, d_model, max_length):
        super(PositionalEncoder, self).__init__()
        self.d_model = d_model
        self.max_length = max_length
        
        # Initialize the positional encoding matrix
        pe = torch.zeros(max_length, d_model)

        position = torch.arange(0, max_seq_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2, dtype=torch.float) * -(math.log(10000.0) / d_model))
        
        # Calculate and assign position encodings to the matrix
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)
    
    # Update the embeddings tensor adding the positional encodings
    def forward(self, x):
        x = x + self.pe[:, :x.size(1)]
        return x
```

```python
Daily XP
802
Exercise
Exercise
Setting up a multi-headed attention class
In multi-headed attention, the outputs from different attention heads are combined together and then adjusted to maintain the same size for further processing.

Let's implement one now starting with the __init__() method.

Instructions
Set the number of attention heads using num_heads.
Set up the linear transformations for the attention inputs: query, key, and value.


class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super(MultiHeadAttention, self).__init__()
        # Set the number of attention heads
        self.num_heads = num_heads
        self.d_model = d_model
        self.head_dim = d_model // num_heads
		# Set up the linear transformations
        self.query_linear = nn.Linear(d_model, d_model)
        self.key_linear = nn.Linear(d_model, d_model)
        self.value_linear = nn.Linear(d_model, d_model)      
        self.output_linear = nn.Linear(d_model, d_model)
```

<h1>Build Encoder Transformer (incorporates self attention mechanism)</h1>
<p>Has 2 main components: Transformer Body and Transformer Head</p>
<p>Transformer Body: stack of multiple encoder layers learning complex patterns in the language data</p>
<p>- each encoder layer has a multi-head self-attention mechanism to capture relationships between elements in the sequence</p>
<p>- feed-forward (sub)layers to map knowlege to abstract non linear representations</p>
<p>Transformer Head: process encoded inputs (hidden states) to produce output prediction</p>
<p>&nbsp;</p>
<p>Feed-forward sublayers in encoding layers</p>
<p>- will contain 2 fully connected linear layers and relu activation&nbsp;</p>
<p>- d_ff: dimension between linear layers to further facilitate capturing complex patterns</p>
<p>- forward(): performs the forward pass through this component using the attention mechanism as the input</p>
<p>&nbsp;</p>
<p>Encoder Layer:</p>
<p>- 2 key elements for implementing an encoder layer</p>
<p>- Multi-headed self-attention and feed-forward sublayer</p>
<p>- incorporated into the encoder layer. together for layer normalization for keeping consistant scales</p>
<p>- dropout to regularize and stablize training</p>
<p>- mask is used during a forward pass to prevent padding tokens (ensures equal length)</p>
<p>&nbsp;</p>

<p>&nbsp;</p>

```python
Post-attention feed-forward layer
Let's assemble some of the pieces of an encoder transformer, starting with the feed-forward sublayer that follows multi-headed self-attention in every encoder layer.

Instructions
Specify in the __init__() method the sizes of the two linear fully connected layers.
Apply a forward pass through the two linear layers, using the ReLU() activation in between.

class FeedForwardSubLayer(nn.Module):
    # Specify the two linear layers' input and output sizes
    def __init__(self, d_model, d_ff):
        super(FeedForwardSubLayer, self).__init__()
        self.fc1 = nn.Linear(d_model, d_ff)
        self.fc2 = nn.Linear(d_ff, d_model)
        self.relu = nn.ReLU()
	
    # Apply a forward pass 
    def forward(self, x):
        return self.fc2(self.relu(self.fc1(x)))
```

```python
Time for an encoder layer
You've made it quite far in building your own skeleton transformer architecture! Now you are ready to assemble a full encoder layer containing:

A multi-headed self-attention mechanism.
A feed-forward sublayer.
A combined layer normalization and dropout to be applied after each of the above two stages.
Instructions
100 XP
Complete the implementation of the EncoderLayer class to initialize all its inner elements one by one.

# Complete the initialization of elements in the encoder layer
class EncoderLayer(nn.Module):
    def __init__(self, d_model, num_heads, d_ff, dropout):
        super(EncoderLayer, self).__init__()
        self.self_attn = MultiHeadAttention(d_model, num_heads)
        self.feed_forward = FeedForwardSubLayer(d_model, d_ff)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, mask):
        attn_output = self.self_attn(x, x, x, mask)
        x = self.norm1(x + self.dropout(attn_output))
        ff_output = self.feed_forward(x)
        return self.norm2(x + self.dropout(ff_output))
```

```python
Encoder transformer body and head
Almost there! Now that the encoder layer implementation has been completed, all that remains is:

Implementing the transformer body, namely a stack of multiple encoder layers.
Appending a task-specific transformer head to process the encoder's resulting hidden states and produce the final outputs for the language task at hand!
Instructions 1/2
50 XP
1
2
Define a stack of multiple encoder layers in the __init__() method.
Complete the forward() method. Note that the process starts by converting the original sequence tokens in x into embeddings.

class TransformerEncoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_layers, num_heads, d_ff, dropout, max_sequence_length):
        super(TransformerEncoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.positional_encoding = PositionalEncoder(d_model, max_sequence_length)
        # Define a stack of multiple encoder layers
        self.layers = nn.ModuleList([EncoderLayer(d_model, num_heads, d_ff, dropout) for _ in range(num_layers)])
	
    # Complete the forward pass method
    def forward(self, x, mask):
        x = self.embedding(x)
        x = self.positional_encoding(x)
        for layer in self.layers:
            x = layer(x, mask)
        return x
    

    
    
Add final linear layer to project encoder results into raw classification outputs.
Apply the necessary function to map raw classification outputs into log class probabilities.

class TransformerEncoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_layers, num_heads, d_ff, dropout, max_sequence_length):
        super(TransformerEncoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.positional_encoding = PositionalEncoder(d_model, max_sequence_length)
        # Define a stack of multiple encoder layers
        self.layers = nn.ModuleList([EncoderLayer(d_model, num_heads, d_ff, dropout) for _ in range(num_layers)])
	
    # Complete the forward pass method
    def forward(self, x, mask):
        x = self.embedding(x)
        x = self.positional_encoding(x)
        for layer in self.layers:
            x = layer(x, mask)
        return x

class ClassifierHead(nn.Module):
    def __init__(self, d_model, num_classes):
        super(ClassifierHead, self).__init__()
        # Add linear layer for multiple-class classification
        self.fc = nn.Linear(d_model, num_classes)

    def forward(self, x):
        logits = self.fc(x[:, 0, :])
        # Obtain log class probabilities upon raw outputs
        return F.log_softmax(logits, dim=-1)
    
```

<h1>Building a decoder transformer (generating text)</h1>
<p>- autoregressive sequence tasks like text generation and completion</p>
<p>- masked multi-head self-attention: helps model predict next word in sequence one step at a time</p>
<p>- for each token in target sequence, only previous token are observed. subsequent tokens are hidden using a upper triangular mask</p>
<p>&nbsp;</p>
<p>Decoder-only transformed head</p>
<p>- estimate likelyhood of each word to be generated</p>
<p>&nbsp;</p>
<p>Mask self-attention</p>
<p>- autoregressive or self behavior</p>
<p>- triangular (casual) attention mask</p>
<p>- token only pays attention to past ('left hand side' in corr plot) information in a sequence</p>
<p>- during training, the token 'favorite', in the sequence, ' orange is my favorite fruit'. Model will focus on the past words 'ornage is my favorite'</p>

```python
Building a decoder body and head
Time to design a high-level architecture for a decoder-only transformer! On this occasion, instead of building the model body and the model head in two separate classes, the model head will be incorporated as part of the model body class that contains the stack of decoder layers.

As usual, the necessary imports for this exercise have been done for you.

class TransformerDecoder(nn.Module):
    def __init__(self, vocab_size, d_model, num_layers, num_heads, d_ff, dropout, max_sequence_length):
        super(TransformerDecoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.positional_encoding = PositionalEncoding(d_model, max_sequence_length)
        self.layers = nn.ModuleList([DecoderLayer(d_model, num_heads, d_ff, dropout) for _ in range(num_layers)])

        # Add a linear layer (head) for next-word prediction
        self.fc = nn.Linear(d_model, vocab_size)

    def forward(self, x, self_mask):
        x = self.embedding(x)
        x = self.positional_encoding(x)
        for layer in self.layers:
            x = layer(x, self_mask)

        # Apply the forward pass through the model head
        x = self.fc(x)
        return F.log_softmax(x, dim=-1)
```
